```python
# Encoding 
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
#Let us split our dataset into x and y
x=data.drop('class',axis=1)
y=data['class']
```

```python
#Let us split train and test data set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

# Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
Gaussian_nb=GaussianNB()
Gaussian_nb.fit(x_train,y_train)
y_nb_pred=Gaussian_nb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
print("Accuracy is: ", accuracy_score(y_test,y_nb_pred))
print("Precision is: ", precision_score(y_test,y_nb_pred, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_nb_pred, average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_nb_pred, average = 'micro'))
```

```python
from sklearn.naive_bayes import BernoulliNB
bernoulli_nb=BernoulliNB()
bernoulli_nb.fit(x_train,y_train)
y_bernoulli_nb_pred=bernoulli_nb.predict(x_test)
```

```python
print("Accuracy is: ", accuracy_score(y_test,y_bernoulli_nb_pred))
print("Precision is: ", precision_score(y_test,y_bernoulli_nb_pred, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_bernoulli_nb_pred, average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_bernoulli_nb_pred, average = 'micro'))
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print("Accuracy for: ",lr, accuracy_score(y_test,y_pred))
print("Precision is: ",lr, precision_score(y_test,y_pred))
print("Recall is: ",lr, recall_score(y_test,y_pred))
print("F1 score is: ",lr, f1_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print("Accuracy for: ",knn, accuracy_score(y_test,y_pred))
print("Precision is: ",knn, precision_score(y_test,y_pred))
print("Recall is: ",knn, recall_score(y_test,y_pred))
print("F1 score is: ",knn, f1_score(y_test,y_pred))
```

# Conclusion

From the above algorithms it is clear that KNeighbors and Logistic regression gave has good accuracy, precision score, recall and F1 score.
