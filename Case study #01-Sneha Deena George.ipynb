```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading dataset 
data=pd.read_csv("cars_data (1).csv")
data.head()
```

```python
#Displaying all columns
columns = data[['Make','Model','Nickname','Car Gender','Buyer Gender','Buyer Age','Buzzword','Country','City','Dealer Latitude','Dealer Longitude','Color','New Car','Purchase Date','Sale Price','Discount','Resell Price','5-yr Depreciation','Top Speed','0-60 Time']]
```

```python
#checking for null values
boolseries = columns.isna()
boolseries.head()
```

```python
#plot a bar graph of male vs female buyers participated in the sales
column1 = data['Buyer Gender']
fig = plt.figure(figsize = (5,5))
plt.bar(column1, color ='maroon', width = 0.4, height = 5000)
plt.xlabel("Buyer Gender")
plt.title("male vs female buyers participated in sales")
plt.show()
```

```python
#top 5 cars based on their sales price
data.sort_values("Sale Price", ascending = False, inplace = True)
data.head()
```

```python
#least 5 cars based on their Resell price
data.nsmallest(5, ['Resell Price'])
```
